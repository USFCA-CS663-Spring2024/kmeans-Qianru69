Compare my KMeans implementation with scikit-learn KMeans

<strong>1. Generate test dataset</strong>

```python
from sklearn.datasets import make_blobs

X, cluster_assignments = make_blobs(n_samples=700, centers=4, cluster_std=0.60, random_state=0)
X
```

```python
cluster_assignments
```

Here I generated 700 data points in 2D space, and these data points belongs to one of 4 clusters. 
* "X" is the array of the data points
* "Cluster_assignments" is the ground truth of the data clustering.

<strong>2. Import my K-Means and perform clustring on X</strong>

```python
from cluster import *
my_cluster_predict = cluster(4)
my_cluster_labels, centroids = my_cluster_predict.fit(X)
print(my_cluster_labels)
centroids
```

The clustering result on the input data X was generated by using my implementation of K-Means algorithm. 
* "my_cluster_labels" is the clustering result using my K-Means algorithm (balanced = False).
* "centroids" is the center value of each cluster.

<strong>3. Import scikit-learn KMeans and perform clustering on X</strong>

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

The clustering result on the input data X was generated by using KMeans implemented in scikit-learn.
* "kmeans.labels_" is the clustering result using scikit-learn K-Means algorithm.
* "kmeans.cluster_centers_" is the center value of each cluster.

<strong>4. Compare my_cluster_labels with the ground truth</strong>

```python
from sklearn import metrics

# Compare against expected labels
rand_score = metrics.rand_score(cluster_assignments, my_cluster_labels)
adjusted_rand = metrics.adjusted_rand_score(cluster_assignments, my_cluster_labels)
print("Rand Index comparing my labels to true labels:", adjusted_rand)
print("Adjusted Rand Index comparing my labels to true labels:", adjusted_rand)
```

Rand index is a method that compares the similarity of two clustering assignments. Perfect labeling is scored 1.0. Poor labeling will show low scores (adjusted rand score close to 0).     
Based on the above result, the Rand Index and adjusted Rand Index scores are close to 1.0, indicating good performance with my K-Means algorithm.    
Reference: https://scikit-learn.org/stable/modules/clustering.html#clustering-performance-evaluation

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

df = pd.DataFrame(X, columns=['Column 1', 'Column 2'])  # Convert X into a DataFrame
df['My Cluster Labels'] = my_cluster_labels  # Add my_cluster_labels into df
df['True Cluster Assignments'] = cluster_assignments  # # Add ground truth labels into df

plt.figure(figsize=(20, 6))

# Plot 1: My Cluster Labels
plt.subplot(1, 2, 1)  # 1st subplot
sns.scatterplot(data=df, x='Column 1', y='Column 2', hue='My Cluster Labels', s=50, alpha=0.5)
plt.title('My Cluster Labels') 
plt.legend(title='Label')

# Plot 2: True Cluster Assignments
plt.subplot(1, 2, 2)  # 2nd subplot
sns.scatterplot(data=df, x='Column 1', y='Column 2', hue='True Cluster Assignments', s=50, alpha=0.5)
plt.title('True Cluster Assignments')  
plt.legend(title='Label')

plt.show()
```

From the above plots, it is easy to observe that the cluster labels between "My Cluster Labels" and "True Cluster Assignments" are very similar. One of the data points labeled differently is at around [0.4, 2.8].

### <strong>KMeans clustering analysis of Chicago Taxi data</strong>

* Reasons why select KMeans:
  * K-Means clustering is a popular unsupervised machine learning algorithm. 
  * KMeans is suitable for large datasets as it has high computational efficiency.

* Perform elbow analysis on inertia values to find the best number of clusters
* Perform KMeans clustering and show the clustering plots

```python
# Perform elbow test to find the best number of clusters based on the pickup locations
inertias = pd.DataFrame(data=[], index=range(2, 10), columns=['intertia'])
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x)
    inertias.loc[k] = kmeans.inertia_
```

```python
inertias
```

```python
inertias.plot()
```

Here I use inertia analysis to find the best number of clusters to in KMeans analysis for the Chicago Taxi dataset. I select 4 as the number of clusters in the following KMeans analysis.

```python
# Label data into 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(x)
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(20, 6)) 
# First subplot with full x and y dimensions
plt.subplot(1, 2, 1)  # 1st subplot
sns.scatterplot(data=df[pickup_columns], x='Pickup Centroid Latitude', y='Pickup Centroid Longitude', hue=kmeans.labels_, s=40, alpha=1, palette='Set1')
plt.title('Pickup coordinate clustering (full dimention)')
plt.legend(title='Label')

# Second subplot with limited x and y dimensions
plt.subplot(1, 2, 2)  # 2nd subplot
sns.scatterplot(data=df[pickup_columns], x='Pickup Centroid Latitude', y='Pickup Centroid Longitude', hue=kmeans.labels_, s=40, alpha=1, palette='Set1')
plt.xlim(41.6, 42.1)  # Set the x-axis range
plt.ylim(-87.5, -87.95)  # Set the y-axis range
plt.title('Pickup coordinate clustering (Zoom in)')
plt.legend(title='Label')

plt.show()
```

From the above plots, we can see that all missing coordinate values (set as 0.0) were clustered into the same one. The rest datapoints were clustered into three clusters.

<strong>Evaluate the clustering output of Chicago Taxi dataset from KMeans</strong>

```python
silhouette_avg = metrics.silhouette_score(x, kmeans.labels_)
print(f"The average silhouette score is : {silhouette_avg}")
```

Here I use the Silhouette Coefficient to evaluate the performance of KMeans clustering output, as for this clustering analysis, we do not have the ground truth. 
* The result shows a Silhouette Coefficient score of 0.71, which is considered good. A score close to 1 indicates the data points are well placed within their cluster, while a score close to -1 indicates that the points are poorly clustered.
* The result is as expected.

### <strong>Hierarchical clustering analysis of Mopsi data subset</strong>

* Reasons why select Hierachical:
  * The geometric locations in this dataset form irregular shapes.
  * Based on the shape in the scatter plot, there might be some hierachical relationship within the data points


```python
from sklearn.cluster import AgglomerativeClustering

k = 5   # set up the same number of clustering as KMeans
metric = 'euclidean'
linkage = 'ward'
hierachical_cluster = AgglomerativeClustering(n_clusters=k, metric=metric, linkage=linkage)
hierachical_cluster.fit(x)
# Association between observations & clusters.
hierachical_cluster.labels_
```

```python
sns.scatterplot(data=df, x='Latitude', y='Longitude', hue=hierachical_cluster.labels_, s=40, alpha=1, palette='Set1')
plt.title('Mopsi data subset clustering by Hierachical clustering')
plt.legend(title='Label')
plt.show()
```

Cluster Label result by Hierarchical clustering is shown in the above scatter plot. The result is similat to KMeans clustering result.

```python
from scipy.cluster.hierarchy import linkage, dendrogram 

Z = linkage(x, 'ward') # Z is now the clustered data
# Graph the data using matplotlib
plt.figure(figsize=(50, 2))
dendrogram(Z)
plt.show()
```

Cluster Label result by Hierarchical clustering presented in a dendrogram

<strong>Evaluate the clustering ouput of Mopsi data subset from Hierarchical clustering</strong>

```python
silhouette_avg = metrics.silhouette_score(x, hierachical_cluster.labels_)
print(f"The average silhouette score is : {silhouette_avg}")
```

Here I use the Silhouette Coefficient to evaluate the performance of Hierarchical clustering output, as for this clustering analysis, we do not have the ground truth.

* The result shows a Silhouette Coefficient score of 0.72.
* The Silhouette Coefficient is similar to (slightly lower than) the KMeans clustering result and is higher than the DBSCAN clustering result.
* The result is as expected.
